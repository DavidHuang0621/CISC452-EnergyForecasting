```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Load processed data
df = pd.read_csv('../Data/preprocessed.csv', low_memory=False, index_col=0)
cols = df.columns
df.info()
```

```python
cols = ['temperature', 'humidity', 'visibility',
       'apparentTemperature', 'pressure', 'windSpeed',
       'windBearing', 'precipIntensity', 'dewPoint', 'precipProbability']
```

```python
from matplotlib import pyplot as plt
fig, axs = plt.subplots(5, 2, figsize=(10,15))
for ax, c in zip(axs.flat, cols):
    
    # Draw the density plot
    ax.set_title(c)
    ax.hist(df[c], bins=25)
    ax.set_xlabel('unit')
    ax.set_ylabel('frequency')
    
# Plot formatting
plt.tight_layout()
plt.show()
```
